```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display


df = pd.read_csv('/content/Extended_Employee_Performance_and_Productivity_Data.csv')


def avaliar_funcionario(row):
    score = 0
    reuniao_necessaria = False
    treinamento_necessario = False


    if row['Performance_Score'] >= 4:
        score += 3
    elif row['Performance_Score'] == 3:
        score += 1
        treinamento_necessario = True
    else:
        score -= 3
        reuniao_necessaria = True


    if row['Employee_Satisfaction_Score'] >= 8:
        score += 2
    elif 6 <= row['Employee_Satisfaction_Score'] <= 7:
        treinamento_necessario = True
    elif row['Employee_Satisfaction_Score'] <= 5:
        score -= 2
        reuniao_necessaria = True


    if row['Overtime_Hours'] > 10:
        score -= 1
        reuniao_necessaria = True
    if row['Sick_Days'] > 5:
        score -= 1
        reuniao_necessaria = True


    if row['Remote_Work_Frequency'] >= 3:
        score -= 1


    if row['Years_At_Company'] > 5:
        score += 1
    if row['Promotions'] > 0:
        score += 1


    if score >= 2:
        return 'Manter'
    elif treinamento_necessario:
        return 'Treinar'
    elif reuniao_necessaria:
        return 'Reunião'
    else:
        return 'Demitir'


def consultar_funcionario(emp_id):
    if emp_id.isdigit():
        emp_id = int(emp_id)
        funcionario = df[df['Employee_ID'] == emp_id]

        if not funcionario.empty:
            resultado = avaliar_funcionario(funcionario.iloc[0])
            output_single.value = f"Funcionário {emp_id}: {resultado}"
        else:
            output_single.value = "Funcionário não encontrado."
    else:
        output_single.value = "ID inválido. Insira um número."


def listar_todos():
    manter_list = []
    demitir_list = []
    treinar_list = []

    for index, row in df.iterrows():
        resultado = avaliar_funcionario(row)
        if resultado == 'Manter':
            manter_list.append(row['Employee_ID'])
        elif resultado == 'Demitir':
            demitir_list.append(row['Employee_ID'])
        elif resultado == 'Treinar':
            treinar_list.append(row['Employee_ID'])


    manter_output.value = f"Funcionários a Manter: {manter_list}" if manter_list else "Nenhum funcionário a manter."
    demitir_output.value = f"Funcionários a Demitir: {demitir_list}" if demitir_list else "Nenhum funcionário a demitir."
    treinar_output.value = f"Funcionários a Treinar: {treinar_list}" if treinar_list else "Nenhum funcionário a treinar."


input_box = widgets.Text(placeholder='Insira o Employee_ID')
button_single = widgets.Button(description="Consultar")
output_single = widgets.Label()


button_single.on_click(lambda b: consultar_funcionario(input_box.value))


listar_button = widgets.Button(description="Listar Todos")
manter_output = widgets.Label()
demitir_output = widgets.Label()
treinar_output = widgets.Label()


listar_button.on_click(lambda b: listar_todos())


tab_single = widgets.VBox([input_box, button_single, output_single])
tab_all = widgets.VBox([listar_button, manter_output, demitir_output, treinar_output])


tab = widgets.Tab(children=[tab_single, tab_all])
tab.set_title(0, 'Consulta Individual')
tab.set_title(1, 'Listar Todos')


display(tab)
```

```python
import cv2
import pickle

img = cv2.imread('estacionamento.png')

vagas = []

# Serve para salvar todas as vagas em um só objeto para gravar.
for x in range(69):
    vaga = cv2.selectROI('vagas', img, False)
    cv2.destroyWindow('vagas')
    vagas.append(vaga)

# Este for serve pra selecionar cada uma das vagas com um retângulo.
for x, y, w, h in vagas:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Este bloco serve pra salvar o arquivo fisicamente com a extensão pkl (vagastotal.pkl)
with open('vagastotal.pkl', 'wb') as arquivo:
    pickle.dump(vagas, arquivo)
```
